```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.page_load_strategy = 'normal'
driver = webdriver.Chrome(options=options)
```

```python
driver.get("https://www.cricbuzz.com/cricket-series/5945/indian-premier-league-2023/matches")
```

```python
dates = []
teams = []
links = []
winners = []
times = []
for i in range(3, 81):
    try:
        date_xpath = "//*[@id='series-matches']/div[" + str(i) + "]/div[1]/span[1]"
        date = driver.find_element(By.XPATH, date_xpath).text
        dates.append(date)
    except Exception as e:
        dates.append('')
    
    try:
        time_xpath =  "//*[@id='series-matches']/div[" + str(i) + "]/div[3]/div[2]/span"
        time = driver.find_element(By.XPATH, time_xpath).text
        times.append(time)
    except Exception as e:
        times.append('')
    
    try:
        team_xpath = "//*[@id='series-matches']/div[" + str(i) + "]/div[3]/div[1]/a[1]/span"
        team = driver.find_element(By.XPATH, team_xpath).text
        teams.append(team)
    except Exception as e:
        teams.append('')
    
    try:
        winner_xpath = "//*[@id='series-matches']/div[" + str(i) + "]/div[3]/div[1]/a[2]"
        winner = driver.find_element(By.XPATH, winner_xpath).text
        winners.append(winner)
    except Exception as e:
        winners.append('')
    
    try:
        link_xpath = "//*[@id='series-matches']/div[" + str(i) + "]/div[3]/div[1]/a[1]"
        link = driver.find_element(By.XPATH, link_xpath).get_attribute("href")
        links.append(link)
    except Exception as e:
        links.append('')
```

```python
data1 = {
    'Date': dates,
    'Time': times,
    'Teams': teams,
    'Winners': winners,
    'Links': links
}
df1 = pd.DataFrame(data1)
df1
```

```python
df1.to_csv('df1_2008.csv')
```
