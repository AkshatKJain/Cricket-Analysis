```python
import pandas as pd
import requests
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.page_load_strategy = 'normal'
driver = webdriver.Chrome(options=options)
```

```python
def get_links(url):
    
    driver.get(url)

    root_div1 = driver.find_element(By.XPATH, "//*[@id='main-container']/div[5]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[1]")
    child_divs1 = root_div1.find_elements(By.XPATH, "./div")

    no_of_matches = len(child_divs1) + 1
    links = []

    for i in range(1, no_of_matches):
        link_xpath = "//*[@id='main-container']/div[5]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[1]/div[" + str(i) + "]/div[1]/div[2]/a"
        link = driver.find_element(By.XPATH, link_xpath).get_attribute("href")
        links.append(link)
    
    modified_links = [link.replace('full-scorecard', 'match-overs-comparison') for link in links]
    
    return modified_links
```

```python
def over_by_over_summary(modified_links, year):
    
    details = []
    team1s = []
    team2s = []
    team1_total_scores = []
    team2_total_scores = []
    winning_details = []
    overs = []
    runs_after_overs = []
    runs_in_overs = []
    batting_teams = []
    bowling_teams = []
    years = []

    for link in modified_links:
        driver.get(link)

        try:
            detail_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]"
            detail = driver.find_element(By.XPATH, detail_xpath).text
        except Exception as e:
            print("skipped at detail")
            continue

        try:
            team1_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/a[1]/span[1]"
            team1 = driver.find_element(By.XPATH, team1_xpath).text
        except Exception as e:
            print("skipped at team1")
            continue

        try:
            team2_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/div[2]/div[1]/a[1]/span[1]"
            team2 = driver.find_element(By.XPATH, team2_xpath).text
        except Exception as e:
            print("skipped at team2")
            continue

        try:
            team1_total_score_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[2]/strong[1]"
            team1_total_score = driver.find_element(By.XPATH, team1_total_score_xpath).text
        except Exception as e:
            print("skipped at team1_total_score")
            continue

        try:
            team2_total_score_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/div[2]/div[2]/strong[1]"
            team2_total_score = driver.find_element(By.XPATH, team2_total_score_xpath).text
        except Exception as e:
            print("skipped at team2_total_score")
            continue

        try:
            winning_detail_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/p[1]/span[1]"
            winning_detail = driver.find_element(By.XPATH, winning_detail_xpath).text
        except Exception as e:
            print("skipped at winning_detail")
            continue

        for i in range(1, 21):
            years.append(year)
            details.append(detail)
            team1s.append(team1)
            team2s.append(team2)
            batting_teams.append(team1)
            bowling_teams.append(team2)
            team1_total_scores.append(team1_total_score)
            team2_total_scores.append(team2_total_score)
            winning_details.append(winning_detail)

            try:
                over_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[2]/div[1]/div[2]/table/tbody/tr[" + str(i) + "]/td[1]/div[1]"
                over = driver.find_element(By.XPATH, over_xpath).text
                overs.append(over)
            except Exception as e:
                overs.append('')

            try:
                runs_after_over_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[2]/div[1]/div[2]/table/tbody/tr[" + str(i) + "]/td[2]/div[1]/div[1]/div[1]"
                runs_after_over = driver.find_element(By.XPATH, runs_after_over_xpath).text
                runs_after_overs.append(runs_after_over)
            except Exception as e:
                runs_after_overs.append('')

            try:
                runs_in_over_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[2]/div[1]/div[2]/table/tbody/tr[" + str(i) + "]/td[2]/div[1]/div[1]/div[1]/span[1]"
                runs_in_over = driver.find_element(By.XPATH, runs_in_over_xpath).text
                runs_in_overs.append(runs_in_over)
            except Exception as e:
                runs_in_overs.append('')


        for i in range(1, 21):
            years.append(year)
            details.append(detail)
            team1s.append(team1)
            team2s.append(team2)
            batting_teams.append(team2)
            bowling_teams.append(team1)
            team1_total_scores.append(team1_total_score)
            team2_total_scores.append(team2_total_score)
            winning_details.append(winning_detail)

            try:
                over_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[2]/div[1]/div[2]/table/tbody/tr[" + str(i) + "]/td[1]/div[1]"
                over = driver.find_element(By.XPATH, over_xpath).text
                overs.append(over)
            except Exception as e:
                overs.append('')

            try:
                runs_after_over_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[2]/div[1]/div[2]/table/tbody/tr[" + str(i) + "]/td[3]/div[1]/div[1]/div[1]"
                runs_after_over = driver.find_element(By.XPATH, runs_after_over_xpath).text
                runs_after_overs.append(runs_after_over)
            except Exception as e:
                runs_after_overs.append('')

            try:
                runs_in_over_xpath = "//*[@id='main-container']/div[5]/div[1]/div[1]/div[3]/div[1]/div[2]/div[1]/div[2]/table/tbody/tr[" + str(i) + "]/td[3]/div[1]/div[1]/div[1]/span[1]"
                runs_in_over = driver.find_element(By.XPATH, runs_in_over_xpath).text
                runs_in_overs.append(runs_in_over)
            except Exception as e:
                runs_in_overs.append('')



    data = {
        'Year' : years,
        'Match Detail' : details,
        'Team 1': team1s,
        'Team 2': team2s,
        'Total Score for Team 1' : team1_total_scores,
        'Total Score for Team 2' : team2_total_scores,
        'Winning Details' : winning_details,
        'Batting team': batting_teams,
        'Bowling team' : bowling_teams,
        'Over Number' : overs,
        'Runs Scored in over' : runs_in_overs,
        'Runs Scored till that over' : runs_after_overs
    }

    df = pd.DataFrame(data)
    
    return df
```

```python
world_cup_links = {
    2023 : "https://www.espncricinfo.com/series/indian-premier-league-2023-1345038/match-schedule-fixtures-and-results"
    2022 : "https://www.espncricinfo.com/series/indian-premier-league-2022-1298423/match-schedule-fixtures-and-results"
    2021 : "https://www.espncricinfo.com/series/ipl-2021-1249214/match-schedule-fixtures-and-results"
    2019 : "https://www.espncricinfo.com/series/ipl-2019-1165643/match-schedule-fixtures-and-results",
    2018 : "https://www.espncricinfo.com/series/ipl-2018-1131611/match-schedule-fixtures-and-results",
    2017 : "https://www.espncricinfo.com/series/ipl-2017-1078425/match-schedule-fixtures-and-results",
    2016 : "https://www.espncricinfo.com/series/ipl-2016-968923/match-schedule-fixtures-and-results",
    2015 : "https://www.espncricinfo.com/series/pepsi-indian-premier-league-2015-791129/match-schedule-fixtures-and-results",
    2014 : "https://www.espncricinfo.com/series/pepsi-indian-premier-league-2014-695871/match-schedule-fixtures-and-results",
    2013 : "https://www.espncricinfo.com/series/indian-premier-league-2013-586733/match-schedule-fixtures-and-results",
}
```

```python
df = pd.DataFrame()
for year, link in world_cup_links.items():
    modified_links = get_links(link)
    dftemp = over_by_over_summary(modified_links, year)
    df = pd.concat([df, dftemp], ignore_index = True)
    df
```

```python
df.to_csv("df.csv")
```
